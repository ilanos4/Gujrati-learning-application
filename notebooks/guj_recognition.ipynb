```python
# Load the trained model
model = SimpleCNN()
model.load_state_dict(torch.load('trained_model.pth', weights_only=True))
model.eval()
```

```python
transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor(),
])
```

```python
import os

# Check the contents of the main data directory
print(os.listdir('../data/Gujrati-learning-application'))
```

```python
# Check the contents of the preprocessed directory
print(os.listdir('../data/Gujrati-learning-application/gujarati_character_recognition/data/preprocessed'))
```

```python
# Function to predict
def predict(image_path):
    image = Image.open(image_path).convert('L')  # Convert to grayscale
    image = transform(image).unsqueeze(0)  # Add batch dimension
    with torch.no_grad():
        output = model(image)
        _, predicted = torch.max(output.data, 1)
    return predicted.item()
# Test the prediction function
image_path = '../data/Gujrati-learning-application/gujarati_character_recognition/data/preprocessed/image_25042021_204759.jpg'  # Adjusted path
print(f'Predicted class: {predict(image_path)}')
```
