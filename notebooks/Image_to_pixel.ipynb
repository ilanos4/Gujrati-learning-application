```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image_path = '../data/Gujrati-learning-application/gujarati_character_recognition/data/preprocessed/14.JPG.jpg'
```

```python
from PIL import Image

image = Image.open(image_path)
```

```python
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# Convert to grayscale if necessary
grayscale_image = image.convert('L')  # Convert to grayscale
pixel_values = np.array(grayscale_image)
```

```python
print("Pixel values:")
print(pixel_values)
```

```python
np.savetxt('pixel_values.txt', pixel_values, fmt='%d')
```
